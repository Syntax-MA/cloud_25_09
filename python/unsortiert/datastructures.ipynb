```python
# list

avengers = ["Ironman", "Thor", "Camptain America", "X-men"]

print(avengers)
print(avengers[0])

avengers[2] = "Superman"

print(avengers)

# methoden

print(len(avengers))
#print(dir(avengers))
#print(help(avengers))

avengers.append("Son Goku") # fügt ans Ende an
avengers.append("Hello Kitty")
avengers.insert(1, "Batman") # Einfugen an Index X alle anderen Werte dannach werden um 1 verschoben
avengers.pop() # löscht default den letzten Eintrag, es kann aber auch der index angegeben werden
avengers.remove("Superman") # löscht den ersten passenden Eintrag

avengers.sort() # Alphabetisch sortieren
avengers.reverse() # umkehren der Reihenfolge
avengers.clear()

print(avengers)
```

```python
# set

avengers = {"Ironman", "Thor", "Camptain America"}

print(avengers)

avengers.add("Super Mario")
avengers.add("Thor")
avengers.discard("Aquaman") # -> kein Fehler wenn nicht vorhanden


print(len(avengers)) # englisch für length -> deutsch länge

print(avengers)

print("Aquaman" in avengers)

```

```python
single_list = [7, 2, 9, 4]

double_list = []

for single_number in single_list:

    double_number = single_number * 2
    double_list.append(double_number)

print(double_list)
```

```python
single_list = [7, 2, 9, 4]

#                berechnung | for single_number in | original list
double_list = [single_number*2 for single_number in single_list]

print(double_list)
```

```python
single_list = [1, 2, 3, 4, 5, 6]

double_list = []

for single_number in single_list:
    
    if single_number % 2 == 0:

        double_number = single_number * 2
        double_list.append(double_number)

print(double_list)
```

```python
single_list = [1, 2, 3, 4, 5, 6]

#                berechnung | for single_number in | original list | Bedingung
double_list = [single_number*2 for single_number in single_list if single_number % 2 == 0]

print(double_list)
```
