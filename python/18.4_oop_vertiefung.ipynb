```python
# Objekt ist wie Variable die sowohl Werte(Attribute) 
# als auch Funktionen(Methoden) beinhalten kann.
# Der Baubplan für ein Objekt heißt Class 
# Mit einem Bauplan können mehrere Objekte erstellt werden.

class Auto:

    counter = 0

    def __init__(self, marke="", farbe="", speed=0):
        self.marke = marke 
        self.farbe = farbe
        self.speed = speed
        Auto.counter += 1

    def hupen(self):
        print("huuup")

    def accel(self):
        self.speed +=1
        print("speed is now:", self.speed)


### Class Ende ###

corsa = Auto("vw", "gelb")
punto = Auto()

punto.accel()
punto.accel()
punto.accel()


print("start")
print(punto.marke)
print("stop")

```

```python
class Auto:
    def __init__(self, marke="vw", max_speed=150):
        self.marke = marke
        self.max_speed = max_speed
        #
    def fahren(self):
        print("brum brum")
###

class Sport(Auto):
    def __init__(self, marke="vw", max_speed=250, turbo=True):
        super().__init__(marke, max_speed)
        self.turbo = turbo
    
    def fahren(self):                       # !
        super().fahren()
        print("brum brum brum")
###

class Van(Auto):
    def __init__(self, marke="vw", max_speed=150, schiebetuer_offen=False):
        super().__init__(marke, max_speed)
        self.schiebetuer_offen = schiebetuer_offen

    def schiebetuer_oeffnen(self):
        print("beep")
        self.schiebetuer_offen=True

    def schiebetuer_schliesen(self):
        print("beep")
        self.schiebetuer_offen=False

###

polo = Auto()
golf_gti = Sport()
bulli = Van("vw", 150, True)


print(f"Hier ist ein brandneuer {polo.marke}. Er kann {polo.max_speed} fahren")
print(f"Hier ist ein brandneuer {golf_gti.marke}. Er kann {golf_gti.max_speed} fahren")


golf_gti.fahren()
```
