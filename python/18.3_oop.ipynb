```python
# Bei Objekten muss man sich merken:
# Es gibt immer eine Schablone die Class
# Es kann aber beliebig viele Instanzen/Objekte der Class geben.

# Class = die Schablone
class Auto:
    # Klassenattribut dies ist der Schablone selbst zugeordnet
    # kann später von allen Instanzen aus erreicht werden
    anzahl_auto = 0             # Auto nach Hause telefonieren

    # Konstruktor
    # Ab hier ist die Anleitung wie jedes einzelne Objekt gebaut werden soll.
    # Hier werden die Attribute(quasi Untervariablen) gesetzt
    # Jedes Objekt wird somit für sich diese Variablen haben.
    # Die Unterschiedlichen Objekte später sind so isoliert voneinander.
    def __init__(self, marke_input="ford", farbe_input="schwarz", sonderausstattung=["Klima","ABS"], scheibenwischer=False, speed=0): #Konstruktor
        
        # Instanzattribute:
        self.marke = marke_input
        self.farbe = farbe_input
        self.sonderausstattung = sonderausstattung
        self.scheibenwischer = scheibenwischer
        self.speed = speed
        # Bonus, erhöhe Klassenatribut um 1
        Auto.anzahl_auto += 1
    #
    # Methoden
    # Alle Instanzen/Objecte werden mit diesen Funktionen ausgestattet.
    #
    def hupen(self):
        print("möööp")
    def wischen(self, an_aus):
        self.scheibenwischer = an_aus
    def beschleunige(self, up=1):
        self.speed += up
    def bremse(self, down=1):
        self.speed -= down


###
# Hier endet die Class, es wird anhand der Einrückung erkannt in Python
# Der Teil oberhalb ist eigenständig noch garnicht im Output zu sehen.
# Erst im unteren Abschnitt muss nun die Schablone genutzt werden um Objekte zu erstellen.
###


auto_alex = Auto("Suzuki", "blau")
print(auto_alex.marke, auto_alex.farbe, auto_alex.sonderausstattung[0])

auto_alex.hupen() # -> print("möööp")
auto_alex.hupen() # -> print("möööp")

print("Der Wischer ist auf an:", auto_alex.scheibenwischer)
auto_alex.wischen(True)
print("Der Wischer ist auf an:", auto_alex.scheibenwischer)

#for merkmal in auto_alex.sonderausstattung:
#    print(merkmal)

auto_daniel = Auto("Bentley", "Racing Green")
print(auto_daniel.marke, auto_daniel.farbe)

auto_daniel.hupen()
print(auto_alex.anzahl_auto)
```

```python
# Bei Objekten muss man sich merken:
# Es gibt immer eine Schablone die Class
# Es kann aber beliebig viele Instanzen/Objekte der Class geben.

# Class = die Schablone
class Auto:
    # Klassenattribut dies ist der Schablone selbst zugeordnet
    # kann später von allen Instanzen aus erreicht werden
    anzahl_auto = 0             # Auto nach Hause telefonieren

    # Konstruktor
    # Ab hier ist die Anleitung wie jedes einzelne Objekt gebaut werden soll.
    # Hier werden die Attribute(quasi Untervariablen) gesetzt
    # Jedes Objekt wird somit für sich diese Variablen haben.
    # Die Unterschiedlichen Objekte später sind so isoliert voneinander.
    def __init__(self, marke_input="ford", farbe_input="schwarz", sonderausstattung=["Klima","ABS"], scheibenwischer=False, speed=0): #Konstruktor
        
        # Instanzattribute:
        self.marke = marke_input
        self.farbe = farbe_input
        self.sonderausstattung = sonderausstattung
        self.scheibenwischer = scheibenwischer
        self.speed = speed
        # Bonus, erhöhe Klassenatribut um 1
        Auto.anzahl_auto += 1
    #
    # Methoden
    # Alle Instanzen/Objecte werden mit diesen Funktionen ausgestattet.
    #
    def hupen(self):
        print("möööp")
    def wischen(self, an_aus):
        self.scheibenwischer = an_aus
    def beschleunige(self, up=1):
        self.speed += up
    def bremse(self, down=1):
        self.speed -= down


###
# Hier endet die Class, es wird anhand der Einrückung erkannt in Python
# Der Teil oberhalb ist eigenständig noch garnicht im Output zu sehen.
# Erst im unteren Abschnitt muss nun die Schablone genutzt werden um Objekte zu erstellen.
###


auto_alex = Auto("Suzuki", "blau")
auto_daniel = Auto("Bentley", "Racing Green")
auto_hassan = Auto("Renault", "blau")

### lets race

auto_alex.beschleunige()
auto_daniel.beschleunige(2)
auto_hassan.beschleunige(7)

# wer ist der schnellste?

print(auto_alex.speed)
print(auto_daniel.speed)
print(auto_hassan.speed)

speedy = max(auto_alex.speed, auto_daniel.speed, auto_hassan.speed)

print("Der schnellste ist:", speedy)

# Klassen Instanzen Attribute und Methoden
```
